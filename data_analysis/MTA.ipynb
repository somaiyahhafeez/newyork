# loading data

```python
import pandas as pd
df=pd.read_csv("MTA.csv")
df
```

# cleaning

```python
df['datetime'] = pd.to_datetime(df['transit_timestamp'])
df['date'] = df['datetime'].dt.date
df
```

```python
import datetime
march = df[df['date'] == datetime.date(2025, 3, 24)]
march
```

```python
columns_to_drop = ["transit_timestamp", "date","transit_mode", "station_complex_id", "payment_method", "fare_class_category", "transfers"]
march = march.drop(columns=columns_to_drop)
march
```

```python
march['station_complex'].unique()
```

```python
commutes=march[march['station_complex'].str.contains('125 St|Jay St-MetroTech|34 St-Penn', na=False)]
commutes
```

```python
commutes['hour'] = commutes['datetime'].dt.hour  
commutes.groupby(['station_complex', 'hour']).size() 

```

```python
stations = [
    "125 St (A,C,B,D)",
    "59 St-Columbus Circle (A,B,C,D,1)",
    "7 Av (E,B,D)",
    "47-50 Sts-Rockefeller Ctr (B,D,F,M)",
    "Bryant Pk (B,D,F,M)/5 Av (7)",
    "34 St-Herald Sq (B,D,F,M,N,Q,R,W)"
]

commute_coords = march[
    (march['datetime'].dt.time >= pd.to_datetime("08:00").time()) &
    (march['datetime'].dt.time < pd.to_datetime("09:00").time()) &
    (march['station_complex'].isin(stations))
][['station_complex', 'latitude', 'longitude', 'ridership']]
```

```python
commute_coords = commute_coords.groupby(
    ['station_complex', 'latitude', 'longitude']
)['ridership'].sum().reset_index()
```

```python
print(commute_coords)
```

## for Tuesday

```python
import datetime
tuesday = df[df['date'] == datetime.date(2025, 3, 25)]
tuesday
```

```python
columns_to_drop = ["transit_timestamp", "date","transit_mode", "station_complex_id", "payment_method", "fare_class_category", "transfers"]
tuesday = tuesday.drop(columns=columns_to_drop)
tuesday
```

```python
emina_school_route = [
    "125 St (A,C,B,D)",
    "59 St-Columbus Circle (A,B,C,D,1)",
    "42 St-Port Authority Bus Terminal (A,C,E)",
    "34 St-Penn Station (A,C,E)",
    "14 St (A,C,E)/8 Av (L)",
    "W 4 St-Wash Sq (A,C,E,B,D,F,M)",
    "Canal St (A,C,E)",
    "Chambers St (A,C)/WTC (E)/Park Pl (2,3)/Cortlandt (R,W)",
    "Fulton St (A,C,J,Z,2,3,4,5)",
    "High St (A,C)",
    "Jay St-MetroTech (A,C,F,R)"
]

commute_coords = tuesday[
    (tuesday['datetime'].dt.time >= pd.to_datetime("08:00").time()) &
    (tuesday['datetime'].dt.time <= pd.to_datetime("09:00").time()) &
    (tuesday['station_complex'].isin(emina_school_route))
][['station_complex', 'latitude', 'longitude', 'ridership', 'datetime']]
commute_coords = commute_coords.groupby(
    ['station_complex', 'latitude', 'longitude']
)['ridership'].sum().reset_index()
commute_coords
```
