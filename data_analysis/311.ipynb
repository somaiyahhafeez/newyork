```python
import pandas as pd
df=pd.read_csv("311.csv")
df
```

```python
df.columns
```

```python
df=df[["Created Date", "Closed Date", "Complaint Type", "Incident Zip", "Incident Address", 'Latitude', 'Longitude']]
df
```

```python
df["Created Date"] = pd.to_datetime(df["Created Date"], errors='coerce')
df
```

```python
df["hour"] = df["Created Date"].dt.hour
df
```

```python
df["Complaint Type"] = df["Complaint Type"].replace(
    to_replace=r"^Noise.*", value="Noise", regex=True
)
```

```python
df
```

```python
df["Complaint Type"].unique()
```

```python
df["Category"] = "Other"  

df.loc[df["Complaint Type"].str.contains(r"^Noise", case=False, na=False), "Category"] = "Noise"
df.loc[df["Complaint Type"].str.contains(r"HEAT|HOT WATER|Boilers", case=False, na=False), "Category"] = "Heat"
df.loc[df["Complaint Type"].str.contains(r"Illegal Parking|Blocked Driveway|Derelict|Abandoned Vehicle", case=False, na=False), "Category"] = "Parking"
df.loc[df["Complaint Type"].str.contains(r"Dirty|UNSANITARY|Rodent|Sewer|Food|Sidewalk", case=False, na=False), "Category"] = "Sanitation"
df.loc[df["Complaint Type"].str.contains(r"Scaffold|SAFETY|Elevator", case=False, na=False), "Category"] = "Safety"
df.loc[df["Complaint Type"].str.contains("WATER LEAK|Water System|ELECTRIC|PLUMBING", case=False), "Category"] = "Infrastructure"
df.loc[df["Complaint Type"].str.contains("DOOR|WINDOW|PAINT|FLOORING|General Construction", case=False), "Category"] = "Building"
df.loc[df["Complaint Type"].str.contains("For Hire|Street Condition|Vendor Enforcement|Dumping|Bike Rack", case=False), "Category"] = "Mobility"
df.loc[df["Complaint Type"].str.contains("Consumer|GENERAL|Lost Property|School|Missed Collection|Non-Emergency|Residential Disposal", case=False), "Category"] = "Civic"

heatmap_data = df.groupby(["Incident Zip", "Category"]).size().unstack(fill_value=0)
heatmap_data
```
